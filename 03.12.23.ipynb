```python
import nltk 
from nltk.corpus import stopwords
nltk.download('stopwords')
s_w=stopwords.words('english')
```

```python
cleaned_tokens=[]
for token in tokens:
    if token not in s_w:
        cleaned_tokens.append(token)
print(cleaned_tokens)
```

```python
# stemming: rule based method: - ing  at the end remove,s ----> remove,
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
tokens=[]
for token in cleaned_tokens:
    tokens.append(stemmer.stem(token))
print(tokens)
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
tokens=[]
for token in cleaned_tokens:
    tokens.append(lemmatizer.lemmatize(token))
" ".join(tokens)
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer =WordNetLemmatizer()
s_w=stopwords.words('english')
from nltk.stem import PorterStemmer
stemmer =PorterStemmer()
tokenizer=RegexpTokenizer(r'\w+')
rows_token =[]
for article in df['Article'].values:
    article=article.lower()
    tokens=tokenizer.tokenize(article)
    cleaned_tokens=[]
    for token in tokens:
        if token not in s_w:
            #stem =stemmer.stem(token)
            #cleaned_tokens.append(stem)
            lemma=lemmatizer.lemmatize(token)
            cleaned_tokens.append(lemma)
    rows_token.append(cleaned_tokens)
df['Cleaned_text']=[" ".join(token) for token in rows_token]
df
```
