```python
import pandas as pd
data={
    'year':[2023,2022,2021],
    'month':[3,4,5],
    'day':[16,17,18]
}
df=pd.DataFrame(data)
df=pd.to_datetime(df)
df.dt.strftime('%m/%d/%y')
```

```python
date=['2023-12-25','2023-12-26']
date=pd.to_datetime(date,format='%Y-%m-%d')
for d in date:
    print(d.strftime('%m/%d/%Y'))
```

```python
# take dob of 5 user as input from the console and convert that to datetime and store in datafram
date=[ '19/11/1994',
 '16/09/1999',
 '19/11/1993',
 '01/07/2022',
 '12/12/2000'
]
date=pd.to_datetime(date)
for d in date:
    print(f'{d.strftime('%d/%m/%Y')} having day {d.day_name()}, on month {d.month_name()}')
```
