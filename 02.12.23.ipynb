```python
# inner join or inner merge
import pandas as pd

data1={
    'name':['a','b','c'],
    'age':[50,40,30]
}
data2 ={
    'name':['b','c','d'],
    'qualified':[True,False,True]
}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
df=df1.merge(df2)
df
```

```python
df = df1.merge(df2,how="left")
df
```

```python
df = df1.merge(df2,how="right")
df
```

```python
df = df1.merge(df2,how="outer")
df
```

```python
df2=pd.DataFrame()
df2['Name']=['b','d','e']
df2['balance']=[20,30,40]
df2['email']=['b@gmail.com','d@gmail.com','e@gmail.com']
df2
```

```python
# left join
left_join=df1.merge(df2,how="left")
left_join
```

```python
# right join
right_join=df1.merge(df2,how="right")
right_join
```

```python
# outer join
outer_join=df1.merge(df2,how="outer")
outer_join
```
